```python
from micrograd.engine import Value
```

```python
a = Value(-4)
b = Value(2)
```

```python
a
```

```python
b
```

```python
c = a+b
d = a*b+b**3
c += c+1
c += 1+c+(-a)
d += d*2 + (b+a).relu()
d += 3*d + (b-a).relu()
e = c-d
f = e**2
```

```python
g = f/2.0
g += 10.0/f
print(f'{g.data:.4f}')
```

```python
g.backward()
print(f'{g.data:.4f}')
```

```python
print(f'{a.grad:.4f}')
print(f'{b.grad:.4f}')
```
